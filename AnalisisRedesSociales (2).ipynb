1. Numero de personas por ingreso

```python
import pandas as pd
from googletrans import Translator

#DEFINICIÓN DE LA SERIE: Contar la frecuencia de cada categoría de ingreso.

# Se crea la variable 'personas_por_ingreso' que faltaba.
personas_por_ingreso = df_redesSociales['Ingreso_Mensual_Aproximado'].value_counts()

# Inicializar el traductor
traductor = Translator()

# Función y Proceso de Traducción (Optimizado)
def translate_text_pt_es(text):
    """Traduce un solo texto de portugués ('pt') a español ('es') usando la instancia global."""
    if pd.isna(text) or not isinstance(text, str):
        return text

    try:
        # Usamos la instancia global del traductor
        translated = traductor.translate(text, src='pt', dest='es')
        return translated.text
    except Exception as e:
        # Si la traducción falla por cualquier motivo (conexión, límite), devuelve el original
        return text

# Crear una lista con los valores del los índices traducidos
print("Iniciando traducción de índices...")
indices_traducidos = [translate_text_pt_es(index) for index in personas_por_ingreso.index]

# Se aplican los índices traducidos a una copia de la Serie
personas_por_ingreso_espanol = personas_por_ingreso.copy()
personas_por_ingreso_espanol.index = indices_traducidos

# Mostrar el resultado final
print("\n--- Conteo por Ingreso Mensual (Traducido) ---")
display(personas_por_ingreso_espanol)
```

B): ¿Qué ocupaciones son más propensas a comprar después de ver un anuncio en redes?: transformar los datos de texto (respuestas categóricas) en formatos numéricos que Pandas puede promediar o calcular.

```python
import pandas as pd
import numpy as np

# Preparación y Limpieza de Datos:
print("Iniciando limpieza y normalización de datos (FINAL)...")

# Normalizar la columna de Compra_Tras_Anuncio a valores binarios (1/0)
# Asume 'sim' (sí en portugués) como el valor positivo
df_redesSociales['Compra_Binaria'] = (
    df_redesSociales['Compra_Tras_Anuncio']
    .astype(str)
    .str.strip()
    .str.lower()
    .apply(lambda x: 1 if 'sim' in x or 'sí' in x else 0)
)

# Normalizar las columnas de Influencia/Frecuencia a valores numéricos (Escala Likert)
# Este mapeo es crucial para convertir texto ('pouco', 'muito') a números (0-4).
mapeo_influencia = {
    'nada': 0,
    'pouco' : 1,
    'médio' : 2,
    'muito' : 3,
    'totalmente' : 4,
    'nenhuma': 0,
    'sim' : 4,
    'nao': 0
}

def mapear_influencia(columna):
    return (
        df_redesSociales[columna]
        .astype(str)
        .str.strip()
        .str.lower()
        .map(mapeo_influencia)
        .fillna(np.nan) # np.nan se usa para que los valores no mapeados se ignoren en el promedio
    )

df_redesSociales['Influencia_Algoritmos_Num'] = mapear_influencia('Influencia_Algoritmos_Anuncios')
# Nombre de columna CORREGIDO: 'Anuncios_Personalizados_Aumentan_Compra'
df_redesSociales['Anuncios_Aumentan_Compra_Num'] = mapear_influencia('Anuncios_Personalizados_Aumentan_Compra')

# Limpiar las columnas de agrupación (Ocupación y Edad)
df_redesSociales['Ocupacion_Principal_Limpia'] = df_redesSociales['Ocupacion_Principal'].astype(str).str.strip().str.lower()
df_redesSociales['Rango_Edad_Limpia'] = df_redesSociales['Rango_Edad'].astype(str).str.strip().str.lower()

# ANÁLISIS: Porcentaje de Compra por Ocupación

print("\n--- ANÁLISIS 1: Porcentaje de Compra por Ocupación ---")

df_temp_compra = df_redesSociales.dropna(subset=['Ocupacion_Principal_Limpia'])

porcentaje_compra_por_ocupacion = (
    df_temp_compra.groupby('Ocupacion_Principal_Limpia')['Compra_Binaria']
    .mean() # Calcula el promedio de 1s y 0s (es el porcentaje)
    .mul(100)
    .sort_values(ascending=False)
    .round(2)
)

print("\nPorcentaje (%) de encuestados que compró 'Tras Ver un Anuncio' (Top 10):")
print(porcentaje_compra_por_ocupacion.head(10).astype(str) + '%')
porcentaje_compra_por_ocupacion
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from googletrans import Translator # Necesario si la traducción no se ha definido

# --- 1. DEFINICIONES PREVIAS (Para evitar NameError) ---

# Define la función de traducción y la instancia (si no se han definido antes)
traductor = Translator()

def translate_text_pt_es(text):
    """Traduce un solo texto de portugués ('pt') a español ('es')."""
    if pd.isna(text) or not isinstance(text, str):
        return text
    try:
        translated = traductor.translate(text, src='pt', dest='es')
        return translated.text
    except Exception as e:
        return text

# --- 2. FILTRADO, CONTEO Y TRADUCCIÓN (Lo que faltaba) ---

# 2a. Filtrar el DataFrame para incluir solo a quienes realizaron una compra después de un anuncio
# Usamos .str.strip().str.capitalize() para asegurar que 'Sim' sea el valor correcto
df_redesSociales['Compra_Tras_Anuncio'] = df_redesSociales['Compra_Tras_Anuncio'].str.strip().str.capitalize()
compras_por_anuncio = df_redesSociales[df_redesSociales['Compra_Tras_Anuncio'] == 'Sim']

# 2b. Contar la cantidad de hombres y mujeres que realizaron una compra
compradores_por_sexo = compras_por_anuncio['Sexo'].value_counts()

# 2c. Traducir los índices de la serie 'compradores_por_sexo' a español
print("Traduciendo índices...")
compradores_por_sexo_espanol = compradores_por_sexo.rename(index=lambda x: translate_text_pt_es(x))


# --- 3. CÓDIGO PARA EL GRÁFICO (Tu bloque original) ---
plt.figure(figsize=(7, 5))

# Usamos un gráfico de barras simple para mostrar la frecuencia de cada sexo
ax = sns.barplot(
    x=compradores_por_sexo_espanol.index,
    y=compradores_por_sexo_espanol.values,
    palette="viridis"
)


# 4. Etiquetas y Títulos
plt.title('Compradores por Sexo (Tras Anuncio)', fontsize=16)
plt.xlabel('Sexo', fontsize=12)
plt.ylabel('Cantidad de Compradores', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 5. Añadir Etiquetas de Datos
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 10),
        textcoords='offset points',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- 1. AGRUPAR Y CONTAR (DEFINICIÓN DE df_conteo_tiempo) ---
# Este bloque es necesario para que el DataFrame exista.
print("Creando DataFrame de contingencia...")
df_conteo_tiempo = (
    df_redesSociales.groupby(['Ocupacion_Principal', 'Tiempo_Diario_Redes_Sociales'])
    .size()
    .reset_index(name='Total_Personas')
)

# Ordenar el resultado (opcional, pero mantiene el orden de la tabla)
df_conteo_tiempo = df_conteo_tiempo.sort_values(
    ['Ocupacion_Principal', 'Total_Personas'],
    ascending=[True, False]
)

# --- 2. NORMALIZAR PARA EL GRÁFICO DE BARRAS APILADAS AL 100% ---

# Calcular el porcentaje de cada combinación dentro de cada grupo de Ocupacion_Principal
df_conteo_tiempo['Porcentaje'] = (
    df_conteo_tiempo.groupby('Ocupacion_Principal')['Total_Personas']
    .transform(lambda x: 100 * x / x.sum())
)

# --- 3. ORDENAR CATEGORÍAS Y GENERAR LA GRÁFICA ---

# Definir el orden correcto de las categorías de tiempo (si aún están en portugués)
orden_tiempo = [
    'Menos de 1 hora',
    '1 a 3 horas',
    '3 a 5 horas',
    'Mais de 5 horas' # Revisa si este texto es correcto, puede ser 'Mais de 5h'
]

plt.figure(figsize=(12, 7))

sns.barplot(
    x='Ocupacion_Principal',
    y='Porcentaje',
    hue='Tiempo_Diario_Redes_Sociales',
    data=df_conteo_tiempo,
    hue_order=orden_tiempo,
    palette="flare"
)

# Etiquetas y Títulos
plt.title('Distribución del Tiempo en Redes por Ocupación (Normalizado al 100%)', fontsize=16)
plt.xlabel('Ocupación Principal', fontsize=12)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.legend(title='Tiempo Diario', bbox_to_anchor=(1.05, 1), loc='upper left')

# Rotar el eje X para mejorar la lectura de las ocupaciones
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
