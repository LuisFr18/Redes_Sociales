```python
import pandas as pd
from googletrans import Translator

# 1. ⚠️ DEFINICIÓN DE LA SERIE: Contar la frecuencia de cada categoría de ingreso.
# Esto crea la variable 'personas_por_ingreso' que faltaba.
personas_por_ingreso = df_redesSociales['Ingreso_Mensual_Aproximado'].value_counts()

# 2. Inicializar el traductor una sola vez
traductor = Translator()

# 3. Función y Proceso de Traducción (Optimizado)
def translate_text_pt_es(text):
    """Traduce un solo texto de portugués ('pt') a español ('es') usando la instancia global."""
    if pd.isna(text) or not isinstance(text, str):
        return text

    try:
        # Usamos la instancia global del traductor
        translated = traductor.translate(text, src='pt', dest='es')
        return translated.text
    except Exception as e:
        # Si la traducción falla por cualquier motivo (conexión, límite), devuelve el original
        return text

# Crear una lista con los valores del índice traducidos
print("Iniciando traducción de índices...")
indices_traducidos = [translate_text_pt_es(index) for index in personas_por_ingreso.index]

# Aplicar los índices traducidos a una copia de la Serie
personas_por_ingreso_espanol = personas_por_ingreso.copy()
personas_por_ingreso_espanol.index = indices_traducidos

# 4. Mostrar el resultado
print("\n--- Conteo por Ingreso Mensual (Traducido) ---")
display(personas_por_ingreso_espanol)
```

B): ¿Qué ocupaciones son más propensas a comprar después de ver un anuncio en redes?: transformar los datos de texto (respuestas categóricas) en formatos numéricos que Pandas puede promediar o calcular.

```python
import pandas as pd
import numpy as np

# Suponiendo que df_redesSociales ya está cargado con tus datos

# --- PARTE A: Preparación y Limpieza de Datos ---
print("Iniciando limpieza y normalización de datos (FINAL)...")

# 1. Normalizar la columna de Compra_Tras_Anuncio a valores binarios (1/0)
# Asume 'sim' (sí en portugués) como el valor positivo
df_redesSociales['Compra_Binaria'] = (
    df_redesSociales['Compra_Tras_Anuncio']
    .astype(str)
    .str.strip()
    .str.lower()
    .apply(lambda x: 1 if 'sim' in x or 'sí' in x else 0)
)

# 2. Normalizar las columnas de Influencia/Frecuencia a valores numéricos (Escala Likert)
# Este mapeo es crucial para convertir texto ('pouco', 'muito') a números (0-4).
mapeo_influencia = {
    'nada': 0,
    'pouco' : 1,
    'médio' : 2,
    'muito' : 3,
    'totalmente' : 4,
    'nenhuma': 0,
    'sim' : 4,
    'nao': 0
}

def mapear_influencia(columna):
    return (
        df_redesSociales[columna]
        .astype(str)
        .str.strip()
        .str.lower()
        .map(mapeo_influencia)
        .fillna(np.nan) # np.nan se usa para que los valores no mapeados se ignoren en el promedio
    )

df_redesSociales['Influencia_Algoritmos_Num'] = mapear_influencia('Influencia_Algoritmos_Anuncios')
# Nombre de columna CORREGIDO: 'Anuncios_Personalizados_Aumentan_Compra'
df_redesSociales['Anuncios_Aumentan_Compra_Num'] = mapear_influencia('Anuncios_Personalizados_Aumentan_Compra')

# 3. Limpiar las columnas de agrupación (Ocupación y Edad)
df_redesSociales['Ocupacion_Principal_Limpia'] = df_redesSociales['Ocupacion_Principal'].astype(str).str.strip().str.lower()
df_redesSociales['Rango_Edad_Limpia'] = df_redesSociales['Rango_Edad'].astype(str).str.strip().str.lower()


# ----------------------------------------------------------------------
## ANÁLISIS 1: Porcentaje de Compra por Ocupación
# ----------------------------------------------------------------------

print("\n--- ANÁLISIS 1: Porcentaje de Compra por Ocupación ---")

df_temp_compra = df_redesSociales.dropna(subset=['Ocupacion_Principal_Limpia'])

porcentaje_compra_por_ocupacion = (
    df_temp_compra.groupby('Ocupacion_Principal_Limpia')['Compra_Binaria']
    .mean() # Calcula el promedio de 1s y 0s (es el porcentaje)
    .mul(100)
    .sort_values(ascending=False)
    .round(2)
)

print("\nPorcentaje (%) de encuestados que compró 'Tras Ver un Anuncio' (Top 10):")
print(porcentaje_compra_por_ocupacion.head(10).astype(str) + '%')
porcentaje_compra_por_ocupacion
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Preparación: Ordenar la serie de mayor a menor frecuencia
# El método sort_values() ordena la Serie de Pandas.
datos_ordenados = personas_por_ingreso_espanol.sort_values(ascending=False)

# 2. Configuración del Gráfico
plt.figure(figsize=(10, 6)) # Define el tamaño de la figura (ancho, alto)

# Usamos seaborn (aunque plt.bar también funciona) para un mejor estilo visual
ax = sns.barplot(
    x=datos_ordenados.index,      # Eje X: Los rangos de ingresos (índices traducidos)
    y=datos_ordenados.values,     # Eje Y: El conteo de personas (los valores de la Serie)
    palette="viridis"             # Define un esquema de color
)

# 3. Etiquetas y Títulos
plt.title('Distribución de Encuestados por Rango de Ingreso Mensual', fontsize=16)
plt.xlabel('Rango de Ingreso Mensual (Salarios Mínimos)', fontsize=12)
plt.ylabel('Cantidad de Encuestados', fontsize=12)

# Rotar las etiquetas del eje X para mejor legibilidad, especialmente para las categorías combinadas
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añadir líneas de cuadrícula para mejor lectura

# 4. Añadir Etiquetas de Datos (Etiquetas de Conteo sobre cada barra)
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}', # Formato del número (entero)
        (p.get_x() + p.get_width() / 2., p.get_height()), # Posición X y Y
        ha='center', va='center',
        xytext=(0, 10), # Desplazamiento vertical (10 puntos arriba)
        textcoords='offset points',
        fontsize=10
    )

plt.tight_layout() # Ajusta automáticamente los parámetros para dar espacio a las etiquetas
plt.show() # Muestra el gráfico
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de tu Serie (Estudiante: 25)
# Recreamos la Serie para asegurar que el código sea reproducible
datos_estudiantes = pd.Series([25], index=['Estudiante'])

# 1. Preparación de los datos
etiquetas = datos_estudiantes.index.tolist() # Lista de etiquetas: ['Estudiante']
valores = datos_estudiantes.values.tolist() # Lista de valores: [25]

# 2. Configuración del Gráfico
plt.figure(figsize=(7, 7))

# 'autopct' formatea el texto del porcentaje dentro de la gráfica.
# 'startangle=90' inicia el primer corte desde la parte superior.
plt.pie(
    valores,
    labels=etiquetas,
    autopct='%1.0f%%', # Muestra el porcentaje como un número entero (100%)
    startangle=90,
    colors=['#B22222'], # Color rojo para resaltar
    wedgeprops={'edgecolor': 'black'} # Añade un borde negro
)

plt.title('Conteo de Estudiantes (100% del Subconjunto Seleccionado)', fontsize=14)
plt.axis('equal') # Asegura que el gráfico sea un círculo perfecto
plt.show()
```
